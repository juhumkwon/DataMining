```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import OneClassSVM

#   ( 2D  )
normal_data = np.random.randn(100, 2)

#   (    )
outliers = np.random.randn(10, 2) + 5  #    

# One-Class SVM   (RBF  )
model = OneClassSVM(nu=0.1, kernel="rbf", gamma="auto")

#    
model.fit(normal_data)

#   (  + ) 
test_data = np.vstack([normal_data, outliers])
predictions = model.predict(test_data)

#    
#   'green',  'red'
colors = ['g' if pred == 1 else 'r' for pred in predictions]

# 
plt.figure(figsize=(8, 6))

#       
plt.scatter(test_data[:, 0], test_data[:, 1], c=colors, edgecolors='k', marker='o', s=100)

#     
plt.title('One-Class SVM for Anomaly Detection')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

#  
plt.show()
```
