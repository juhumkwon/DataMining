```python
# 필요한 라이브러리 임포트
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# 2D 데이터 생성 (클러스터를 3개로 설정)
# cluster_std: 각 클러스터의 표준 편차를 설정합니다. 값이 클수록 군집이 넓게 퍼집니다.
X, y = make_blobs(n_samples=300, centers=3, cluster_std=0.60, random_state=0)

# k-평균 군집화 모델 정의 (k=3)
# n_clusters=4: 군집의 개수를 3로 설정합니다.
kmeans = KMeans(n_clusters=3)

# 모델 학습
kmeans.fit(X)

# 군집의 중심점
# centers = kmeans.cluster_centers_는 K-평균 군집화 모델이 학습을 완료한 후, 각 군집의 중심점 (centroid)을 가져오는 코드입니다.
centers = kmeans.cluster_centers_
print("centers = kmeans.cluster_centers_:",centers)

# 예측된 군집 라벨
y_kmeans = kmeans.predict(X)

# 결과 시각화
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')

# 중심점 표시
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.5)
plt.show()
```
