```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import entropy

# 데이터 생성
np.random.seed(42)

# 정상 데이터 (정규분포)
normal_data = np.random.normal(loc=0, scale=1, size=1000)

# 이상치 데이터 (정규분포, 평균이 이동됨)
outlier_data = np.random.normal(loc=3, scale=1, size=1000)

# 히스토그램으로 확률 밀도 함수(PDF) 추정
bins = np.linspace(-4, 8, 50)
normal_hist, _ = np.histogram(normal_data, bins=bins, density=True)
outlier_hist, _ = np.histogram(outlier_data, bins=bins, density=True)

# 엔트로피 계산
entropy_normal = entropy(normal_hist)
entropy_outlier = entropy(outlier_hist)

# KL 다이버전스 계산 (normal -> outlier)
kl_div = entropy(normal_hist, qk=outlier_hist)

# 시각화
plt.figure(figsize=(10, 6))

# 정상 데이터와 이상치 데이터의 분포
plt.hist(normal_data, bins=bins, density=True, alpha=0.5, label="Normal Data (Entropy={:.2f})".format(entropy_normal))
plt.hist(outlier_data, bins=bins, density=True, alpha=0.5, label="Outlier Data (Entropy={:.2f})".format(entropy_outlier))

# 텍스트로 KL 다이버전스 표시
plt.text(5, 0.3, f"KL Divergence = {kl_div:.2f}", fontsize=12, color='red')

# 그래프 설정
plt.title("Information Theoretic Approaches: Entropy and KL Divergence")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.show()
```
