```python
import matplotlib.pyplot as plt

# Number of clusters and corresponding WCSS values
k_values = [1, 2, 3, 4, 5]
wcss_values = [10, 2.5, 1.5, 1.0, 0.8]  # Example values decreasing as k increases

# Create the plot
plt.figure(figsize=(8, 5))
plt.plot(k_values, wcss_values, marker='o', linestyle='-', color='b', label='WCSS')

# Add labels, title, and legend
plt.title("Elbow Method for Optimal k (WCSS vs Number of Clusters)", fontsize=14)
plt.xlabel("Number of Clusters (k)", fontsize=12)
plt.ylabel("WCSS (Within-Cluster Sum of Squares)", fontsize=12)
plt.xticks(k_values)  # Show integer values for k on the x-axis
plt.grid(True, linestyle='--', alpha=0.7)
plt.axvline(x=2, color='r', linestyle='--', label='Optimal k (Elbow Point)')
plt.legend()

# Show the plot
plt.show()
```
